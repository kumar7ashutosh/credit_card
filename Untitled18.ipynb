```python
import numpy as np,pandas as pd,pickle 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
columns_withna=['V9'	,'V10'	,'V11'	,'V12'	,'V13'	,'V14'	,'V15'	,'V16'	,'V17'	,'V18'	,'V19'	,'V20'	,'V21'	,'V22'	,'V23'	,'V24'	,'V25'	,'V26'	,'V27'	,'V28',	'Amount'	,'Class']
for i in columns_withna:
  df[i]=df[i].fillna(df[i].mean())
```

```python
df.iloc[182328,9:]
```

```python
df.isna().sum()
```

```python
actual=df[df.Class<1]
fake=df[df.Class>=1]
```

```python
actual.shape,fake.shape
```

```python
actual_sample=actual.sample(n=492)
```

```python
new_df=pd.concat([actual_sample,fake],axis=0)
```

```python
new_df['Class'].value_counts()
```

```python
x=new_df.drop(columns=['Class'],axis=1)
y=new_df['Class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
x_train_pred=model.predict(x_train)
training_data_acc=accuracy_score(x_train_pred,y_train)
```

```python
training_data_acc
```

```python
x_test_pred=model.predict(x_test)
testing_data_acc=accuracy_score(x_test_pred,y_test)
testing_data_acc
```
